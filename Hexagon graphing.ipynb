```python
import matplotlib.pyplot as plt
from matplotlib.patches import RegularPolygon
import numpy as np

# Parameters for the honeycomb
rows, cols = 5, 5
hex_radius = 1

# Initialize the plot
fig, ax = plt.subplots(figsize=(8, 8))

# Loop to create the honeycomb pattern
for row in range(rows):
    for col in range(cols):
        # Calculate x and y offsets
        x_offset = col * 1.5 * hex_radius
        y_offset = row * np.sqrt(3) * hex_radius + (col % 2) * np.sqrt(3) / 2 * hex_radius
        
        # Add hexagon
        hexagon = RegularPolygon((x_offset, y_offset), numVertices=6, radius=hex_radius,
                                 orientation=np.radians(30), facecolor='none', edgecolor='black')
        ax.add_patch(hexagon)

        # Calculate and plot the vertices
        for i in range(6):  # 6 corners for a hexagon
            angle = np.radians(60 * i)  # Each corner separated by 60 degrees
            x_corner = x_offset + hex_radius * np.cos(angle)
            y_corner = y_offset + hex_radius * np.sin(angle)
            
            # Plot the corner with a value (example: random value)
            value = np.random.randint(1, 10)  # Replace with your own data
            ax.plot(x_corner, y_corner, 'ro')  # Red dots at corners
            ax.text(x_corner, y_corner, f"{value}", fontsize=8, ha='center', va='center')

# Adjust plot limits
ax.set_xlim(-1, cols * 1.5 * hex_radius)
ax.set_ylim(-1, rows * np.sqrt(3) * hex_radius)
ax.set_aspect('equal')

# Display the plot
plt.title('Honeycomb with Values at Corners')
plt.axis('off')
plt.show()
```
